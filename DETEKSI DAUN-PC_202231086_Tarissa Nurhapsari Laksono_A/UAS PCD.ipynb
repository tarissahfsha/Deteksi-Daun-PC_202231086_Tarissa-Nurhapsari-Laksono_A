```python
# Mengimport library
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path = 'buahh.jpg'
image = cv2.imread(image_path)
```

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
```

```python
# Mendefinisikan rentang warna untuk buah (disesuaikan)
lower_brown = np.array([5, 10, 10])
upper_brown = np.array([20, 255, 255])

# Membuat mask untuk buah
mask_fruit = cv2.inRange(hsv, lower_brown, upper_brown)

# Mendefinisikan rentang warna untuk daun (disesuaikan)
lower_green = np.array([35, 20, 50])
upper_green = np.array([85, 255, 255])

# Membuat mask untuk daun
mask_leaf = cv2.inRange(hsv, lower_green, upper_green)

# Membuat gambar segmentasi buah
fruit_segment = np.zeros_like(image_rgb)
fruit_segment[mask_fruit != 0] = image_rgb[mask_fruit != 0]

# Membuat gambar segmentasi daun
leaf_segment = np.zeros_like(image_rgb)
leaf_segment[mask_leaf != 0] = image_rgb[mask_leaf != 0]
```

```python
# Plot hasil
fig, axs = plt.subplots(2, 2, figsize=(20, 15))

# Gambar 1: Original Image
axs[0, 0].imshow(image_rgb, extent=[0, 1000, 700, 0])
axs[0, 0].set_title('Gambar Asli')
axs[0, 0].axis('on')

# Gambar 2: Mask of Fruits
axs[0, 1].imshow(mask_fruit, cmap='gray', extent=[0, 1000, 700, 0])
axs[0, 1].set_title('Mask Sawo')
axs[0, 1].axis('on')

# Gambar 3: Segmented Fruits
axs[1, 0].imshow(fruit_segment, extent=[0, 1000, 700, 0])
axs[1, 0].set_title('Segmentasi Sawo')
axs[1, 0].axis('on')

# Gambar 4: Segmented Leaves
axs[1, 1].imshow(leaf_segment, extent=[0, 1000, 700, 0])
axs[1, 1].set_title('Segmentasi Daun')
axs[1, 1].axis('on')

# Menampilkan plot
plt.tight_layout()
plt.show()
```
